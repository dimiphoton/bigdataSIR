# step 1: setting up the model #

```python
# the lmfit module uises an orderd dict structure
# to store the parameters to be optimized
# https://lmfit.github.io/lmfit-py/parameters.html

params = Parameters()
params.add('beta_SI', value=0.15, min=0.02, max=1)
params.add('gamma_IR', value= 0.05, min=0.02, max=1)


#params.add('S0',value=5,min=1,max=20)
params.add('I0',value=20,min=4,max=200)
#params.add('R0',value=5,min=0,max=20)



params.add('O_nbtest_SI',value=0.25,min=0.05,max=1)
params.add('O_dailyH_IR',value =7.5, min=1.5, max=20)


params.add('Tr_I_H', value=0.02,min=0.001,max=0.1)
params.add('Tr_H_R', value=0.5,min=0.01,max=1)

params.add('P_Hcum_ICUcum',value=0.66, min=0.2,max=1.2)
params.add('P_ICUcum_D',value=0.0002, min=0.00002,max=0.0004)


# params.add('OBS_H_to_ICU',value=0.1,min=0.01,max=0.5) mistake??
```

```python
def deriv(compartiments, t, params):
    S,I,R,H = compartiments
    beta_SI = params['beta_SI'].value
    gamma_IR = params['gamma_IR'].value
    Tr_I_H = params['Tr_I_H'].value
    Tr_H_R = params['Tr_H_R'].value

    #beta = beta_i*(1.1-tau*t) idée de faire régresser beta
    dSdt = -beta_SI*(S*I/(1e6))
    dIdt = beta_SI*(S*I/(1e6)) - gamma_IR*I - Tr_I_H*I
    dRdt = gamma_IR*I + Tr_H_R*H
    dHdt = Tr_I_H*I - Tr_H_R*H
    
    
    return dSdt, dIdt, dRdt, dHdt
```

```python
""" This function generates a dataframe from the data
input: dataframe from data.csv, parameters params
output: a dataframe with compartiment numbers and transitions
"""

def data2comp(params,data):

    result=pd.DataFrame()
    result['Day']=data['Day']
    result['S->I']=params['O_nbtest_SI'].value*data['num_tested']
    result['I->H']=data['num_cumulative_hospitalizations'].diff()
    result['I->H'][0]=0
    result['I->R']=params['O_dailyH_IR']*result['I->H']
    result['I->R'][0]=0
    result['H->R']=data['num_cumulative_hospitalizations'].diff()-data['num_hospitalised'].diff()
    result['H->R'][0]=0
    
    result['I']=params['I0'].value+result['S->I'].cumsum()-result['I->H'].cumsum()
    result['I'][0]=params['I0'].value
    #result['dH']=data['num_cumulative_hospitalizations'].diff()
    result['H']=data['num_hospitalised']
    result['ICU']=data['num_critical']
    result['D']=data['num_fatalities']
    result['R']=result['I->R'].cumsum()+result['H->R'].cumsum()
    result['R'][0]=0
    result['S']=(1e6-result['I']-result['R']-result['H']-result['ICU']-result['D'])-result['S->I'].cumsum()
    #result['Rnumber']=1e6*result['S->I']/(result['S']*result['I->R'])
    
    return result 
    
```

```python
"""
this function creates a dataframe from parameters (data just for timespan)
input: clinical dataframe, parameters
output: a dataframe with compartiment numbers and transitions
"""

def params2comp(params,data):

    result=pd.DataFrame()
    result['Day']=data['Day']
    array=odeint(deriv,[1e6-params['I0'].value,params['I0'].value,0,0],np.linspace(0,data.shape[0]-1,data.shape[0]), args=(params,))
    result['S']=array.transpose()[0]
    result['I']=array.transpose()[1]
    result['R']=array.transpose()[2]
    result['H']=array.transpose()[3]
    result['S->I']=params['beta_SI'].value*(result['S']*result['I']/(1e6))
    result['I->H']=params['Tr_I_H']*result['I']
    result['I->R']=params['gamma_IR'].value*result['I']
    result['H->R']=params['Tr_H_R'].value*result['H']
    cumH=result['H'].cumsum()
    cumICU=params['P_Hcum_ICUcum'].value*cumH
    D=params['P_ICUcum_D'].value*cumICU
    result['ICU']=cumICU.diff()
    result['ICU'][0]=0
    result['D']=D
    #result['Rnumber']=result['S->I']/(result['S']*result['I->R'])
    
    return result
    
```
